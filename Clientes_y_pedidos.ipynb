**Pregunta 3** ¿Cuál es el estado con más clientes?

```python
# abrir archivo e iniciar vars
# evaluar el contenido del archivo
from os.path import exists
import pandas as pd
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')

  df = pd.read_csv(datafile)

  # Contar la frecuencia de cada estado
  state_counts = df["State"].value_counts()

  # Encontrar el estado con más clientes
  most_common_state = state_counts.idxmax()
  num_customers_in_most_common_state = state_counts.max()

  print(f"El estado con más clientes es '{most_common_state}' con {num_customers_in_most_common_state} clientes.")
```
